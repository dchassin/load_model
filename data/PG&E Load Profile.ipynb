# PGE Load Profile Data

PG&E provide access to historical load data by tariff. The data can obtained from their [Load Profile Archive](https://www.pge.com/en_US/for-our-business-partners/retail-energy-suppliers/electric-service-provider-resource-center/mads/mads-load-profiles.page).  This notebook pulls the data for the month of March 2020 and plots the separate and total load.

```python
import pgande
from datetime import datetime

start = '3/1/20'
stop = '3/31/20'
data = pgande.get_loads(datetime.strptime(start,'%m/%d/%y'),
                        datetime.strptime(stop,'%m/%d/%y'))
# drop columns that are redundant
data.drop(columns=["E1","S19","S1M","S1S","S20","S9M"],inplace=True)
print('Columns....',','.join(list(data.columns)))
print('Rows.......',data.shape[0])
```

```python
data.plot(figsize=(12,8),grid=True)
```

```python
data.sum(axis=1).plot(figsize=(12,8),grid=True)
```
